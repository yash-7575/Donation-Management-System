```python
import numpy as np
import matplotlib.pyplot as plt

def kolam_generator(m, n):
    # m: dots per arm, n: number of arms (gcd(m, n) == 1)
    sequence = []
    for k in range(m):
        val = m if k == 0 else (k * n) % m
        sequence.append(val if val != 0 else m)
    dots_matrix = np.tile(sequence, (n, 1)).T  # m rows, n columns
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    # Plot the Kolam
    fig, ax = plt.subplots()
    for i in range(m):
        for j in range(n):
            r = dots_matrix[i][j]
            theta = angles[j]
            x, y = r * np.cos(theta), r * np.sin(theta)
            ax.plot(x, y, 'ko')  # Plot points
            # Connect to next point (circularly)
            next_i = (i + 1) % m
            next_j = (j + 1) % n
            x_next, y_next = dots_matrix[next_i][next_j] * np.cos(angles[next_j]), dots_matrix[next_i][next_j] * np.sin(angles[next_j])
            ax.plot([x, x_next], [y, y_next], 'b-')
    ax.set_aspect('equal')
    plt.show()

# Example usage:
kolam_generator(m=6, n=10)
```
